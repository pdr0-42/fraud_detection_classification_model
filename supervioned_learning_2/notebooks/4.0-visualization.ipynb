# Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/interim/data_cleaned.csv")
```

```python
df.iloc[:,1:13].hist(figsize = (20,20))
df.iloc[:,14:30].hist(figsize = (20,20))
```

```python
plt.figure(figsize = (22,20))
sns.heatmap(df.corr(),annot = True)
plt.show()
```

```python
for column in df.columns:
    plt.hist(df[column], bins=20, color='blue', alpha=0.5)
    plt.title(f'Histogram of variable: {column}')
    plt.xlabel('Values')
    plt.show()
```
